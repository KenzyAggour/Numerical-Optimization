**Batch GD**

```python
m = len(X)
```

```python
def batchgrad(x , y , LR , iters):
  loss = []
  theta0s = []
  theta1s = []
  theta0 = 0
  theta1 = 0
  hypos = []
  prev_cost = float('inf')

  for i in range(iters):
    print("**********iteration",i,"**********\n")
    theta0s.append(theta0)
    theta1s.append(theta1)  #same size this way
    print("theta 0 ", theta0,"\n")
    print("theta 1 ", theta1,"\n")

    h = theta0 + theta1*x
    hypos.append(h)
    print("h ", h,"\n")
    error_vec = h-y
    print("errorvec " ,error_vec,"\n")
    j = np.mean(error_vec**2) / 2
    loss.append(j)
    print("j ",j,"\n")
    d_theta0 = np.sum(error_vec) / m
    d_theta1 = np.sum(error_vec*x)  / m
    gradvec = np.array([d_theta0,d_theta1])
    print("grad vector ", gradvec,"\n")
    gradvecNorm = np.linalg.norm(gradvec)

    print("grad vector norm ", gradvecNorm,"\n")

    if gradvecNorm < 0.001 or abs(prev_cost - j) < 0.001:
      break
    else:
      prev_cost = j
      theta0 = theta0 - LR * d_theta0
      theta1 = theta1 - LR * d_theta1

  return theta0s, theta1s, loss, theta0, theta1 , h , hypos
```

#### Try your function with any input parameterss of your choice.

```python
theta0s , theta1s , loss , theta0, theta1 , h ,hs= batchgrad(X, y, 0.0005, 300)
```

## Stochastic GD

##### Very Important notes:
- <font color='red'>Save your losses and thetas for each iterations (not each epoch).</font>
- <font color='red'>Plot loss vs. iterations (not epochs). This will allow you to see the difference between SGD (stochastic GD) and batch GD.</font>
- <font color='red'>Convergence check should be performed each epoch not each iterations. This allows you to perform more accurate decision. This is because the noisy process could lead to a deciving decision.</font>
- <font color='red'>Your result may be different from the results shown below due to the randomness of the process. As well as, the random shuffle. The results will be different also for two different runs of the algorithm.</font>


```python
def stochgrad(x , y , LR , eps):
  loss = []
  theta0s = []
  theta1s = []
  theta0 = 0
  theta1 = 0
  hypos = []
  prev_cost = float('inf')

  for i in range(eps):
    print("*******************************************************epoch",i,"**********************************************************************\n")
    for k in range(len(shuffX)):
      print("**********iter",k,"****************************")
      theta0s.append(theta0)
      theta1s.append(theta1)  #same size this way
      print("theta 0 ", theta0,"\n")
      print("theta 1 ", theta1,"\n")
      h = theta0 + theta1*x[k]
      hypos.append(h)
      print("h ", h,"\n")
      error_vec = h-y[k]
      print("errorvec " ,error_vec,"\n")
      j = np.mean(error_vec**2) / 2
      loss.append(j)
      print("j ",j,"\n")
      d_theta0 = np.sum(error_vec)
      d_theta1 = np.sum(error_vec*x[k])
      gradvec = np.array([d_theta0,d_theta1])
      print("grad vector ", gradvec,"\n")
      gradvecNorm = np.linalg.norm(gradvec)
      print("grad vector norm ", gradvecNorm,"\n")
      theta0 = theta0 - LR * d_theta0
      theta1 = theta1 - LR * d_theta1

    #comparing current last loss and last loss of prev epoch
    if gradvecNorm < 0.001 or (i != 0 and abs(loss[-51] - loss[-1]) < 0.001):        #if epoch n not = 0 to skip first one
      break

  return theta0s, theta1s, loss, theta0, theta1 , h , hypos, i
```

## Mini-Batch GD

```python
def minibatchgrad(x, y, batch_size , LR, epochs):
    loss = []
    theta0s = []
    theta1s = []
    theta0 = 0
    theta1 = 0
    hypos = []
    prev_cost = float('inf')
    num_batches = len(shuffX) // batch_size

    for epoch in range(epochs):
        print("**********epoch", epoch, "**********\n")

        for i in range(0, m, batch_size):
            print("-----batch-----\n")

            x_batch = x[i:i + batch_size]
            y_batch = y[i:i + batch_size]

            theta0s.append(theta0)
            theta1s.append(theta1)
            print("theta 0 ", theta0, "\n")
            print("theta 1 ", theta1, "\n")

            h = theta0 + theta1 * x_batch
            hypos.append(h)
            print("h ", h, "\n")
            error_vec = h - y_batch
            print("errorvec ", error_vec, "\n")
            j = np.mean(error_vec**2) / 2
            loss.append(j)
            print("j ", j, "\n")
            d_theta0 = np.sum(error_vec) / batch_size
            d_theta1 = np.sum(error_vec * x_batch) / batch_size
            gradvec = np.array([d_theta0, d_theta1])
            print("grad vector ", gradvec, "\n")
            gradvecNorm = np.linalg.norm(gradvec)
            print("grad vector norm ", gradvecNorm, "\n")
            theta0 = theta0 - LR * d_theta0
            theta1 = theta1 - LR * d_theta1

        if gradvecNorm < 0.001 or (epoch != 0 and abs(loss[-1] - loss[ - (num_batches+1)]) < 0.001):
            break

    return theta0s, theta1s, loss, theta0, theta1, h, hypos
```

##### The following results uses batch size=5, alpha = 0.0005 and max. iterations = 300.

```python
theta0s, theta1s, loss, theta0, theta1, h, hypos = minibatchgrad(shuffX, shuffy, 5 , 0.0005, 300)
```

```python
y = theta0 + theta1 *shuffX
```

```python
r2_score(shuffy,y)
```
